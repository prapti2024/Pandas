```python
file_path = 'brandname3word.csv'
data = pd.read_csv(file_path)

# Drop the unnamed column as it is not needed for analysis
data = data.drop(columns=['Unnamed: 0'])

# Define the list of brands
brands_list = [
    "AHAVA", "AMELIORATE", "APIVITA", "ARK Skincare", "Acorelle", "Aesop",
    "Alchimie Forever", "Alpha", "Ambre Solaire", "Antipodes", "Armani",
    "AromaWorks", "Aromatherapy Associates", "Aurelia", "Australian Bodycare",
    "Avant", "Aveda", "Aveeno", "Avene", "BARBER PRO", "Balance Me", "BeautyPro",
    "Benton", "Bioderma", "Bloom", "Bobbi Brown", "Bondi Sands", "Bubble T",
    "Bulldog", "By Terry", "COSRX", "Caudalie", "CeraVe", "Chantecaille",
    "Dermalogica", "Dr. Dennis Gross", "Elemis", "Estée Lauder", "Eucerin",
    "First Aid Beauty", "Garnier", "Kiehls", "La Roche-Posay", "Neutrogena",
    "Nivea", "Olay", "Ordinary", "Philosophy", "REN", "Rodial", "Sukin",
    "The Ordinary", "Weleda", "Zelens", "bareMinerals", "benefit", "ilapothecary",
    "philosophy", "this works", "yes to"
]

# Extract the brand name if it exists in the brands list
def extract_brand_name(brand):
    words = brand.split()
    for length in range(1, 4):
        candidate = ' '.join(words[:length])
        if candidate in brands_list:
            return candidate
    return words[0]

# Apply the function to the 'Brands' column
data['Brand'] = data['Brands'].apply(extract_brand_name)

# Group by the new 'Brand' column and calculate the average price for each brand
average_prices_by_brand = data.groupby('Brand').Price.mean().reset_index()

# Rename columns for clarity
average_prices_by_brand.columns = ['Brand', 'Average Price']

# Save the complete table to a CSV file
average_prices_by_brand.to_csv('average_prices_by_brand.csv', index=False)

# Display the result
print(average_prices_by_brand)
```
